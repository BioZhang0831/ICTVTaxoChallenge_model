```python
import pickle
from sklearn.preprocessing import LabelEncoder
with open('../data/tmp/encoders.pkl', 'rb') as f:
    encoders = pickle.load(f)
```

```python
with open('../data/tmp/pred_data.pkl', 'rb') as f:
    pred_data = pickle.load(f)
pred_data
```

```python
import pandas as pd
raw_df=pd.read_csv('../data/prediction/filtered_raw_classdf.csv')
raw_df
```

```python
#['phylum', 'class', 'order', 'family', 'genus', 'species']
#raw_df['contigId'] + pred_data
pre_df=pd.DataFrame(pred_data, columns=['phylum', 'class', 'order', 'family', 'genus', 'species'],index=raw_df['contigId'])
```

```python
pre_df['phylum'] = encoders['phylum'].inverse_transform(pre_df['phylum'])
pre_df['class'] = encoders['class'].inverse_transform(pre_df['class'])
pre_df['order'] = encoders['order'].inverse_transform(pre_df['order'])
pre_df['family'] = encoders['family'].inverse_transform(pre_df['family'])
pre_df['genus'] = encoders['genus'].inverse_transform(pre_df['genus'])
pre_df['species'] = encoders['species'].inverse_transform(pre_df['species'])
```

```python
pre_df=pre_df.reset_index()
```

```python
import numpy as np
def merge_df(pre_df,raw_df):
    return pd.DataFrame(np.where(raw_df.isna() & ~pre_df.isna(), pre_df, raw_df), columns=raw_df.columns)

final_result_df=merge_df(pre_df,raw_df)
```

```python
final_result_df
```

```python
final_result_df['superkingdom']='Viruses'

new_order = ['contigId','superkingdom','phylum', 'class', 'order', 'family', 'genus', 'species']

final_result_df = final_result_df[new_order]
final_result_df
```

```python
final_result_df.to_csv('../data/prediction/final_classdf.csv',index=False)
```

```python
import pandas as pd
```

```python
#ICTVTaxoChallenge_model/data/prediction/ICTV_RSCU.csv
rscu_df = pd.read_csv('../data/prediction/ICTV_RSCU.csv')
rscu_df
```

```python
#data/prediction/ICTV_TNF.csv
tnf_df = pd.read_csv('../data/prediction/ICTV_TNF.csv')
tnf_df
```

```python
#/data/prediction/final_ICTVChallenge_taxonomy_result.csv
rawtaxonomy_df = pd.read_csv('../data/prediction/final_ICTVChallenge_taxonomy_result.csv')
rawtaxonomy_df
```

```python
tnfdf_Contigs = set(tnf_df['contigId'])
rscudf_Contigs = set(rscu_df['contigId'])
rawclassdf_Contigs = set(rawtaxonomy_df['contigId'])

common_Contigs = tnfdf_Contigs & rscudf_Contigs & rawclassdf_Contigs
len(common_Contigs)
```

```python
filtered_tnfdf = tnf_df[tnf_df['contigId'].isin(common_Contigs)]
filtered_rscudf = rscu_df[rscu_df['contigId'].isin(common_Contigs)]
filtered_rawclassdf = rawtaxonomy_df[rawtaxonomy_df['contigId'].isin(common_Contigs)]
```

```python
hcname=['phylum', 'class', 'order', 'family', 'genus', 'species']
```

```python
raw_clas_dict = {}
for index, row in filtered_rawclassdf.iterrows():
    base_id = row["contigId"]
    taxonomic_levels = []
    score_list = []
    for key in ['phylum', 'class', 'order', 'family', 'genus', 'species']:
        if pd.isna(row[key]):
            taxonomic_levels.append(None)
            score_list.append(0)
        else:
            taxonomic_levels.append(row[key])
    raw_clas_dict[base_id] = taxonomic_levels
print(len(raw_clas_dict))
raw_clas_dict
```

```python
raw_classdf = pd.DataFrame.from_dict(raw_clas_dict, orient='index', columns=hcname)
raw_classdf.index.name = 'contigId'
raw_classdf
```

```python
raw_classdf.to_csv('../data/prediction/filtered_raw_classdf.csv')
```

```python
import pickle
from sklearn.preprocessing import LabelEncoder
with open('../data/tmp/encoders.pkl', 'rb') as f:
    encoders = pickle.load(f)
encoders
```

```python
# phylum        18
# class         43
# order         86
# family       334
# genus       3691
# species    17671
# 
label_columns = ['phylum', 'class', 'order', 'family', 'genus', 'species']
for col in label_columns:
    for index,row in raw_classdf[col].iteritems():
        try:
            raw_classdf[col][index] = encoders[col].transform([row])[0]
        except:
            #print(f'Error: {row} not found in {col}')
            if col == 'phylum':
                raw_classdf[col][index] = 18
            elif col == 'class':
                raw_classdf[col][index] = 43
            elif col == 'order':
                raw_classdf[col][index] = 86
            elif col == 'family':
                raw_classdf[col][index] = 334
            elif col == 'genus':
                raw_classdf[col][index] = 3691
            elif col == 'species':
                raw_classdf[col][index] = 17671

raw_classdf
```

```python
inputdata = pd.merge(filtered_tnfdf, filtered_rscudf, on='contigId')
inputdata = pd.merge(inputdata, raw_classdf, on='contigId')
inputdata
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder, MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA


# 假设 x_features 是一个已经存在的 DataFrame
scaler = StandardScaler()
minmax_scaler = MinMaxScaler()

# 定义特征的列索引
feature1_cols = inputdata.columns[1:65]  # 数值特征1
feature2_cols = inputdata.columns[65:321]  # 数值特征2
feature3_cols =  inputdata.columns[322:327]  # 分类特征

# 创建列转换器
preprocessor = ColumnTransformer(
    transformers=[
        ('f1', scaler, feature1_cols),
        ('f2', minmax_scaler, feature2_cols),
        ('f3', scaler, feature3_cols)  # 放大分类特征
    ])

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
])

# 应用预处理
x_features = pipeline.fit_transform(inputdata)

x_features
```

```python
import pickle
with open('datax_pre.pkl', 'wb') as f:
    pickle.dump(x_features, f)
```

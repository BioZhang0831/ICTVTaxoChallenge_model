```python
import pandas as pd
```

```python
# Read the CSV file
classdf = pd.read_csv('../data/train/virus_taxonomy.csv', sep=',')
classdf.rename(columns={'VirusGENBANKaccession':'contigId'}, inplace=True)
classdf
```

```python
rawclassdf=pd.read_csv('../data/train/add_names.txt', sep='\t')
rawclassdf['#']= rawclassdf["#"].str.replace(r"\.\d+$", "", regex=True)
rawclassdf.rename(columns={'#':'contigId'}, inplace=True)
rawclassdf
```

```python
rscudf=pd.read_csv('../data/train/combine_RSCU.csv', sep=',')
rscudf['id']= rscudf["id"].str.replace(r"\.\d+$", "", regex=True)
rscudf.rename(columns={'id':'contigId'}, inplace=True)
rscudf
```

```python
tnfdf=pd.read_csv('../data/train/virus_final_tnf.csv', sep=',')
tnfdf['Contig']= tnfdf["Contig"].str.replace(r"\.\d+$", "", regex=True)
tnfdf.rename(columns={'Contig':'contigId'}, inplace=True)
tnfdf
```

```python
tnfdf_Contigs = set(tnfdf['contigId'])
rscudf_Contigs = set(rscudf['contigId'])
rawclassdf_Contigs = set(rawclassdf['contigId'])
classdf_Contigs = set(classdf['contigId'])

common_Contigs = tnfdf_Contigs & rscudf_Contigs & rawclassdf_Contigs & classdf_Contigs
print(len(common_Contigs))
```

```python
filtered_tnfdf = tnfdf[tnfdf['contigId'].isin(common_Contigs)]
filtered_rscudf = rscudf[rscudf['contigId'].isin(common_Contigs)]
filtered_rawclassdf = rawclassdf[rawclassdf['contigId'].isin(common_Contigs)]
filtered_classdf = classdf[classdf['contigId'].isin(common_Contigs)]
```

```python
hcname=['phylum', 'class', 'order', 'family', 'genus', 'species']
```

```python
raw_clas_dict = {}
raw_clas_score_dict={}
for index, row in filtered_rawclassdf.iterrows():
    base_id = row["contigId"]
    taxonomic_levels = []
    score_list = []
    for key in ['phylum', 'class', 'order', 'family', 'genus', 'species']:
        if pd.isna(row[key]):
            taxonomic_levels.append(None)
            score_list.append(0)
        else:
            if len(row[key].split(": ")) > 1:
                taxonomic_levels.append(row[key].split(": ")[0])
                score_list.append(float(row[key].split(": ")[1]))
            else:
                taxonomic_levels.append(None)
                score_list.append(0)
    raw_clas_dict[base_id] = taxonomic_levels
    raw_clas_score_dict[base_id] = score_list
print(len(raw_clas_dict))
```

```python
clas_dict = {}
for index, row in filtered_classdf.iterrows():
    base_id = row["contigId"]
    taxonomic_levels = []
    for key in ['Phylum', 'Class', 'Order', 'Family', 'Genus', 'Species']:
        if pd.isna(row[key]):
            taxonomic_levels.append(None)
        else:
            taxonomic_levels.append(row[key])
    clas_dict[base_id] = taxonomic_levels
print(len(clas_dict))
#clas_dict
```

```python
#pip install gensim
from gensim.models import Word2Vec
node_names = list(clas_dict.values())+list(raw_clas_dict.values())
word2vec_model = Word2Vec(sentences=node_names, vector_size=16, window=3, min_count=1, workers=4)
word2vec_model.save("word2vec.model")
```

```python
clas_df = pd.DataFrame.from_dict(clas_dict, orient='index', columns=hcname)
clas_df.index.name = 'contigId'
raw_classdf = pd.DataFrame.from_dict(raw_clas_dict, orient='index', columns=hcname)
raw_classdf.index.name = 'contigId'
raw_classdf
```

```python
clas_df
```

```python
#raw_classdf+clas_df
combined_df = pd.concat([raw_classdf, clas_df], axis=0)
combined_df
```

```python
from sklearn.preprocessing import LabelEncoder
import pickle
label_columns = ['phylum', 'class', 'order', 'family', 'genus', 'species']
encoders = {}  # 用于存储训练好的编码器，以便将来可以反向转换或者新增数据时再对其进行 transform

for col in label_columns:
    le = LabelEncoder()
    combined_df[col] = le.fit_transform(combined_df[col])
    encoders[col] = le

with open('encoders.pkl', 'wb') as f:
    pickle.dump(encoders, f)
```

```python
for col in label_columns:
    clas_df[col] = encoders[col].transform(clas_df[col])
clas_df
```

```python
for col in label_columns:
    raw_classdf[col] = encoders[col].transform(raw_classdf[col])
raw_classdf
```

```python
#filtered_tnfdf/filtered_rscudf/filtered_rawclassdf/filtered_classdf
traindata = pd.merge(filtered_tnfdf, filtered_rscudf, on='contigId')
traindata = pd.merge(traindata, raw_classdf, on='contigId')
traindata = pd.merge(traindata, clas_df, on='contigId')
traindata
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder, MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from category_encoders import TargetEncoder

# 假设 x_features 是一个已经存在的 DataFrame
scaler = StandardScaler()
minmax_scaler = MinMaxScaler()

# 定义特征的列索引
feature1_cols = traindata.columns[1:65]  # 数值特征1
feature2_cols = traindata.columns[65:321]  # 数值特征2
feature3_cols =  traindata.columns[322:327]  # 分类特征

# 创建列转换器
preprocessor = ColumnTransformer(
    transformers=[
        ('f1', scaler, feature1_cols),
        ('f2', minmax_scaler, feature2_cols),
        ('f3', scaler, feature3_cols)  # 放大分类特征
    ])

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
])

# 应用预处理
x_features = pipeline.fit_transform(traindata)

x_features
```

```python
y_label=traindata[traindata.columns[327:333]].to_numpy()
```
